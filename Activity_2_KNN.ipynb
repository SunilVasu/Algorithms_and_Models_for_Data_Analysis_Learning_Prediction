```r
inputData <-read.csv("Data/german_credit_data1.csv",header=T)
#summary(inputData)
inputData
nrow(inputData)
ncol(inputData)
```

```r
#set.seed(126)
#n.points<-1000
#age <- rpois(n.points, 50)
#income<-rnorm(n.points,mean=50,sd=30)
#credit <- sample(0:1, 1000, replace = TRUE)
#is.numeric(credit)
#is.factor(credit)
credit.f<-factor(inputData$Credit.Risks, label=c("high","low"))
data=data.frame("age"=inputData$Age,"income"=inputData$Credit.amount,"rating"=credit.f)

head(data)
data
n.points <- nrow(data)
n.points
```

# Test and training set (20-80%)

```r
sampling.rate <-0.8
num.test.set.labels<- n.points* (1 - sampling.rate)

training<-sample(1:n.points, sampling.rate*n.points, replace=FALSE)
train<-subset(data[training,], select=c(age,income))

testing<-setdiff(1:n.points,training)
test<-subset(data[testing,], select=c(age,income))

cl<- data$rating[training]
#cl
true.labels<-data$rating[testing]

knn(train,test,cl,k=3) # we know class labels for train; Machine, Learn now.

#The number of Iteration to be done is n
n <- 20
mat <- matrix(ncol=2, nrow=n)

#compute and print misclassification rate
for(k in 1:n) {
  #print(k)
  predicted.labels <- knn(train,test,cl,k)
  num.incorrect.labels<-sum(predicted.labels !=true.labels)
  misclassification.rate<-num.incorrect.labels/num.test.set.labels
  #print(misclassification.rate)
  mat[k,] <- c(k, misclassification.rate)

}
#print(mat)
mat <- data.frame(mat)
library(plyr)
rename(mat, c("X1"="K", "X2"="misclassification.rate"))

#now predict using k with lowest mis-classification rate
cat("Test for (57,37)\nResult :")
testcase<-c(22,5951)
knn(train,testcase,cl,k=17)
```
